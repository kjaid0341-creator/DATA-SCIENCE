# ENCAPSULATION

```python
class mirai_class:
  strength = 100
  teachers = ['A', 'B', 'C']
  div = ['alpha', 'beta']
  sub = ['ds', 'web', 'dsa']

  def __init__(self):
    print('class created')

obj = mirai_class()
print(obj.strength)
print(obj.teachers)
print(obj.div)
print(obj.sub)
```

### NEW CLASS EXAMPLE

```python
class student:
    def __init__(self,name,age):
        self.name=name
        self.age=age


a = student('A',20)
print(a.name)
print(a.age)
```

## Using Display Function

```python
class student:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def display(self):
      print('Name', self.name)
      print('Age', self.age)

a = student('A',20)
a.display()
```
